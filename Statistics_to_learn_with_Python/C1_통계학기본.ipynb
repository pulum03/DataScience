1.1 통계학

1.1.1 통계학의 목적(1) : 가지고 있는 데이터의 설명

통계학이 무어냐고 묻는다면 '데이터를 잘 사용하는 방법을 알아내는 학문'

데이터는 2종류: 1. 지금 가지고 있는 데이터, 2. 가지고 있지 않은 데이터

가지고 있는 데이터에 대해선 데이터의 정리와 요악이 목적

가지고 있는 데이터를 정리 요약하기 위해서 사용되는 통계학 : 기술통계 또는 기술통계학

1.1.2 통계학의 목적(2) : 모르는 데이터의 예측과 추측

가지고 있지 않은 미지의 데이터를 추측할 목적으로 사용되는 통계학 : 추측통계 혹은 추측통계학

e.g.e 내일의 매상데이터

이런 미지의 데이터를 추측하기 위해선 데이터의 활용이 필요. 과거의 데이터를 이용해서 추측하고 제안 할 수있음. 일종의 판매예측

"가지고 있는 데이터를 이용해서 모르는 데이터를 추측"하는 것이 통계학의 큰 이유

1.1.3 표본과 모집단

표본이란 현재 가지고 있는 데이터

모집단이란 아직 가지고 있지 않은 모르는 데이터를 포함한 모든 데이터

표본이라는 일부 데이터를 이용해서 모집단이라는 전체 데이터를 분석하는 것이 추측통계의 목적!

1.2 표본을 얻는 과정

1.2.1 확률변수

확률적 법칙에 따라서 변화하는 값 : 확률변수

1.2.2 호수와 낚시의 예

e.g.호수와 그곳에 낚시로 잡은 물고기

작은 호수에는 물고기가 1종류만 살고 있고, 다른 강이나 호수에서 물고기가 유입되지 않는다고 가정.

물고기를 낚는데 쉽고 어려움 차이 x, 낚은 물고기는 다시 호수에 풀어줘서 전체 물고기 숫자 변동 x, 물고기가 다치거나 하는 일 x (즉, 데이터의 변화량 x)

물고기를 1마리 잡는다면, 1마리가 표분. 모집단은 '호수에 있는 전체 물고기'.

모집단에서 표본을 얻는 것: 샘플링

예제에서는 호수 안의 전체 물고기 중 일부(1마리)를 낚아 올려서 샘플링 한것.

낚아올린 물고기의 길이는 소수점 이하 반올림 3센치

1.2.3 모집단에서 표본 얻는 과정

다음날에도 똑같은 호수에서 똑같은 장비로 낚시를 한다고 가정. 이번에도 1마리만 낚음.

이 때, 물고기의 크기는 몇 센치?(호수에 있는 모든 물고기의 크기를 파악하고 있다고 가정, 즉, 다시말해 모집단의 추정을 완벽하게 할 수 있음)

호수 안에는 물고기 5마리가 있고, 몸길이는 (소수점 이하 반올림)

2cm : 1마리

3cm : 1마리

4cm : 1마리

5cm : 1마리

6cm : 1마리

낚시를 해서 1마리를 낚는다는 것은 5마리 중 1마리를 랜덤으로 고로는 것과 같다.

모집단을 완전히 알고 있어도 내일 낚아 올릴 물고기가 5마리 중 어떤 것이 될지는 알 수 없음.

2센치짜리 물고기가 낚일 확률 : 20% 

5센치까지 물고기가 낚일 확률 역시 : 20%

내일 낚일 물고기의 크기를 예측해보라고 하면 '2센치일 확률은 20%'

'틀림없이 20%라고 주장'하는 것은 불가능

20%확률로 2센치 물고기가 낚일수도, 20% 확률로 3센치짜리 물고기가 낚일 수 도 있음.

이처럼 확률적으로 변화하는 값이기 때문에 내일 낚일 물고기의 크기는 '확률변수'

표본을 확률변수로 간주.

1.2.4 실현값

확률변수의 구체적인 값 : 실현값

특별히 '실현값'이라고 부르는 이유는 얻지 못한 데이터가 있다는 것을 명확하게 표시하기 위해서.

낚아올린 물고기가 3센치였을 경우 '3센치'라는 데이터는 실현값.

하지만 호수에는 2센치나 4센치도 있음. 3센치 이외의 다른 물고기가 잡혔을 수도 있다는 것.

1.2.5 샘플링

모집단에서 표본을 얻은 것 : 샘플링 또는 표분추출

호수에서 낚은 물고기가의 몸길이 데이터를 얻는 것이 샘플링.

설문조사에서 조사결과를 얻는 것이 샘플링.

주사위를 굴려서 나온 눈을 기록하는 것이 샘플링.
